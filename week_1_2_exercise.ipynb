## Getting started- Data science tools and packages

### Pandas
* pandas is a Python library for working with structured data.
* It helps us load, clean, organize, analyze, and save data.
* It is especially useful for data in table format, like Excel files, CSV files, or database tables.
* The main object in pandas is called a DataFrame, which is similar to a table with rows and columns.

```python
import pandas as pd
```

#### Read Data

* The main object in pandas is called a DataFrame, which is similar to a table with rows and columns.

The following cell of code loads data from a CSV file into a Pandas DataFrame.

note:
* pd.read_csv(...) reads a CSV file using the Pandas library.
* 'data/diamonds.csv' is the location of the file (change it to the location of your CSV file)
* sep=',' tells Pandas that the values in the file are separated by commas.
* index_col=0 tells Pandas to use the first column in the CSV file as the row index instead of treating it as a regular data column.
* diamonds_df = saves the loaded table in a variable called diamonds_df.

After running it, diamonds_df contains the diamonds dataset and can be used for data analysis.

```python
diamonds_df = pd.read_csv('data/diamonds.csv', sep=',' , index_col=0)
```

* In The following cell, The head() method displays the first rows of a DataFrame.
* Recall that diamonds_df is the DataFrame that contains the diamonds dataset.
* .head(10) tells Pandas to display the first 10 rows of the DataFrame (chane it to display more or fewer rows using a different number).
* This method is commonly used to quickly inspect the data after loading it and verify that it looks correct.

```python
diamonds_df.head(10)
```

### Descriptive statistics

The describe() method generates a statistical summary (Descriptive statistics) of the numeric columns in a DataFrame.

For each numeric column, it displays:

* count – The number of non-missing (NaN) values.
* mean – The average value.
* std – The standard deviation, which measures how spread out the values are.
* min – The smallest value.
* 25% – The first quartile (25% of the values are below this number).
* 50% – The median (the middle value).
* 75% – The third quartile (75% of the values are below this number).
*  max – The largest value.

```python
diamonds_df.describe().round()
```

Use include=['str'] to describe string variables instead numeric

```python
diamonds_df.describe(include=['str'])
```
